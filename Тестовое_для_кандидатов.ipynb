# Задание 1
Необходимо с помощью SQL запроса собрать одну общую таблицу из всех данных витрин, которая должна включать в себя все поля этих витрин.
В качестве ответа предоставить ссылку на вашу копию страницы, где можно увидеть написанный запрос и собранную таблицу. Ссылку вставьте в Word файл "Техническое задание" в соответствующее поле.


_(одна из таблиц называется `group`, её имя конфликтует с зарезервированными ключевым словом SQL, поэтому его нужно будет заключить в двойные кавычки `""`)_


## Решение

```python
%%sql result <<
SELECT
    mngmnt.org_management_rk as "Ключ управления",
    mngmnt.management_nm  as "Название управления",
    product.hit_rk as "Ключ коммуникации с клиентом",
    product.hid as "Идентификатор продукта",
    COALESCE(product.using_flg, 0) as "Флаг начала пользования продуктом",
    queue.queue_id as "Идентификтор очереди",
    queue.queue_desc as "Расшифровка идентификтора очереди",
    result.hit_status_result_id as "Идентификтор результата коммуникации",
    result.hit_status_result_desc as "Расшифровка идентификтора результата",
    horoscope.agent_login as "Логин оператора",
    horoscope.horoscope as "Знак зодиака оператора",
    "group".org_group_rk as "Ключ группы",
    "group".org_management_rk as "Ключ управления",
    "group".group_nm as "Расшифровка ключа группы",
    emp_x_org_gr.employee_rk as "Ключ оператора",
    task.task_rk as "Ключ задания",
    task.task_stage_id as "Идентификтор стадии задания",
    task.source_system_cd as "Код системы источника",
    task.create_dttm as "Дата-время создания задания",
    task.finish_dttm as "Дата-время окончания задания",
    action.hit_rk as "Ключ коммуникации с клиентом",
    action.hit_status_result_id as "Идентификтор результата коммуникации",
    call.finish_dttm as "Дата-время завершения коммуникации",
    call.duratoin_sec as "Длительность коммуникации в секундах"
FROM
    mngmnt
LEFT JOIN "group" ON mngmnt.org_management_rk = "group".org_management_rk
LEFT JOIN emp_x_org_gr ON "group".org_group_rk = emp_x_org_gr.org_group_rk
LEFT JOIN call ON emp_x_org_gr.employee_rk = call.wo_employee_rk
LEFT JOIN horoscope ON call.agent_login = horoscope.agent_login
LEFT JOIN task ON call.wo_task_rk = task.task_rk
LEFT JOIN action ON call.wo_hit_rk = action.hit_rk
LEFT JOIN product ON call.wo_hit_rk = product.hit_rk
LEFT JOIN result ON action.hit_status_result_id = result.hit_status_result_id
LEFT JOIN queue ON call.wo_queue_id = queue.queue_id;
```

# Задание 2
Необходимо посчитать следующие значения:

* Общее количество звонков со статусом «Дозвон, Успешно»;
* Продажная конверсия, по формуле:

  $$ Конверсия = \frac{N_{Дозвон, Успешно}}{N_{Дозвон, Успешно} + N_{Дозвон, Отказ}}$$
* Средняя длительность одной коммуникации, закрытой статусом «Дозвон, Успешно»

Ответы необходимо вписать в соответствующую таблицу в фаиле "Техническое задание"


## Решение

```python
%%sql
SELECT COUNT(*)
FROM call
JOIN action ON call.wo_hit_rk = action.hit_rk
JOIN result ON action.hit_status_result_id = result.hit_status_result_id
WHERE result.hit_status_result_desc = 'Дозвон, Успешно';
```

```python
%%sql
WITH status_counts AS (
    SELECT
        SUM(CASE WHEN result.hit_status_result_desc = 'Дозвон, Успешно' THEN 1 ELSE 0 END) AS success_count,
        SUM(CASE WHEN result.hit_status_result_desc = 'Дозвон, Отказ' THEN 1 ELSE 0 END) AS failure_count
    FROM call
    JOIN action ON call.wo_hit_rk = action.hit_rk
    JOIN result ON action.hit_status_result_id = result.hit_status_result_id
)
SELECT
    success_count,
    failure_count,
    CAST(success_count AS FLOAT) / (success_count + failure_count) AS conversion_rate
FROM status_counts;
```

```python
%%sql
SELECT AVG(CAST(call.duratoin_sec AS FLOAT)) AS avg_duration
FROM call
JOIN action ON call.wo_hit_rk = action.hit_rk
JOIN result ON action.hit_status_result_id = result.hit_status_result_id
WHERE result.hit_status_result_desc = 'Дозвон, Успешно';
```
